```python
import pandas as pd
import matplotlib.pyplot as plt

import pathlib
```

```python
despesas_file = "despesas_contratadas_candidatos_2024_SP.csv"
receitas_file = "receitas_candidatos_2024_SP.csv"
```

```python
data_path = pathlib.Path("../data/")
```

```python
despesas = pd.read_csv(data_path / despesas_file, encoding='latin-1', sep=";", thousands=',')
receitas = pd.read_csv(data_path / receitas_file, encoding='latin-1', sep=";", thousands=',')
```

```python
# Ajustando o gráfico de pizza para melhorar a disposição dos rótulos
top_k = 5

# Assumindo que o DataFrame 'despesas_soma' já esteja definido e ordenado
top_10 = despesas_soma.sort_values(by="VR_DESPESA_CONTRATADA", ascending=False).head(top_k)

# Função para exibir o valor total junto com a porcentagem
def autopct_format(pct, allvalues):
    absolute = int(pct/100.*sum(allvalues))
    return "{:.1f}%\nR${:,.0f}".format(pct, absolute)

# Plotando o gráfico de pizza com ajustes
plt.figure(figsize=(10, 10))  # Aumentando o tamanho da figura
plt.pie(top_10['VR_DESPESA_CONTRATADA'], labels=top_10.index, 
        autopct=lambda pct: autopct_format(pct, top_10['VR_DESPESA_CONTRATADA']), 
        startangle=90, labeldistance=1.1)  # Aumentando a distância dos rótulos
plt.title(f'Top {top_k} Gastos por Candidato')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
candidato = despesas[despesas["NM_CANDIDATO"]=="GUILHERME CASTRO BOULOS"]
```

```python
candidato.columns
```

```python
# Função personalizada para concatenar strings com '-' e somar valores numéricos
def concat_or_sum(x):
    if x.dtype == 'object':  # Verifica se é string
        return '-'.join(x.dropna().unique())  # Concatenar strings únicas com '-'
    else:
        return x.sum()  # Somar valores numéricos

# Aplicando a função no groupby
gastos_candidato = candidato.groupby(by=["DS_ORIGEM_DESPESA"]).agg(concat_or_sum)
```

```python
gastos_candidato =  gastos_candidato.sort_values(by="VR_DESPESA_CONTRATADA", ascending=False)
```
